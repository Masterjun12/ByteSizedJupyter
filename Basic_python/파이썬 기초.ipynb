# 문제2

```python
# 다음 리스트를 출력된 결과처럼 될 수 있도록 리스트 함수를 써서 중간과정을 채우시오
# 각 문제의 순서에 맞춰 답안 작성해야 같은 결과가 나옴

p = [1, 2, 4, 5]
q = [11, 12, 13]

print(p)   # 출력결과 : [1, 2, 4, 5]

#문제1 : p와 q를 합치기

print(p+q)   # 출력결과 : [1, 2, 4, 5, 11, 12, 13]

#문제2 : 값 3 추가
p.append(3)
r = p+q
print(r)   # 출력결과 : [1, 2, 3, 4, 5, 11, 12, 13]

#문제3 : 값 6 추가
p.append(6)
y = p+q
print(y)   # 출력결과 : [1, 2, 3, 4, 5, 6, 11, 12, 13]

#문제4 : 값 4 제거
result = p + q
result.remove(4)
result.sort()
print(result)   # 출력결과 : [1, 2, 3, 5, 6, 11, 12, 13]

#문제5 : 전체 순서를 거꾸로
u = p+q
u.reverse()
print(u)   # 출력결과 : [13, 12, 11, 6, 5, 3, 2, 1]

#문제6 : q의 값을 3회 반복
i = q+q+q
print(i)   # 출력결과 : [11, 12, 13, 11, 12, 13, 11, 12, 13]
```

# 문제4

```python
# 다음 함수 func을 완성하시오.

def func(a,b=2,c=3): # 킬러 문항
    return (a + b + c, a * b * c)


print(func(4, 5, 2))  # 출력결과 : (11, 40)
print(func(4, 5)   )  # 출력결과 : (12, 60)
print(func(4))        # 출력결과 : (9, 24)

p, q = func(5, 6, 7)

print("p:", p)  # 출력결과 : p: 18
print("q:", q)  # 출력결과 : q: 210
```

# 문제7

```python
from ast import Num
import numpy as np

a = [[4,3,2,1],[1,3,2,4],[7,30,15,18],[10,19,12,18]]
b = [[3,4,2,1],[2,3,4,4],[7,6,5,8],[7,9,2,8]]

aa = np.array(a)
bb = np.array(b)
```

```python
# 행렬 aa와 bb의 행렬곱을 구하시오.
ab = np.matmul(a, b)
print(ab)  # 출력 결과 : [[ 39  46  32  40] [ 51  61  32  61] [312 370 245 391] [278 331 192 326]]
```

```python
# 행렬 bb와 aa의 행렬곱을 구하시오.
ba = np.matmul(b, a)
print(ba)   # 출력 결과 : [[ 40 100  56  73] [ 79 211 118 158] [149 341 197 265] [131 260 158 223]]
```

```python
# 1에서 100까지 숫자중 3의 배수의 array을 생성해서 cc에 넣으시오
num = np.arange(1, 101)
mp = num % 3 == 0
cc = num[mp]
print(cc)
```

```python
# 행렬 bb에서 4미만인 값만 3배하시오.
for i in range(len(bb)):
    for j in range(len(bb[i])):
        if bb[i][j] < 4:
            bb[i][j] *= 3
print(bb)  # 출력 결과 : [[9 4 6 3] [6 9 4 4] [7 6 5 8] [7 9 6 8]]
```

```python
# 자료형을 float로 변환하고, 2x8 행으로 변경
cc = np.concatenate([aa, bb])
cc = np.reshape(aa, (2, 8))
print(cc)
# 출력 결과 : [[ 4.  3.  2.  1.  1.  3.  2.  4.] [ 7. 30. 15. 18. 10. 19. 12. 18.]]
```
